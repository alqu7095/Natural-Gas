```
gas2 = gas.drop(gas.index[:3774])

ng = gas2.rename(index=str, columns={"Date": "date", "Price": "ng_price"})

ng.head()
```

```
ung2 = ung.drop(ung.index[:1213])

ung3 = ung2.drop(['Open', 'High', 'Low', 'Adj Close', 'Volume'], axis=1)

ung4 = ung3.rename(index=str, columns={"Date": "date", "Close": "ung_price"})

print(ung4.shape)

ung4.head()
```

```
ugaz2 = ugaz.drop(['Open', 'High', 'Low', 'Adj Close', 'Volume'], axis=1)

ugaz3 = ugaz2.rename(index=str, columns={"Date": "date", "Close": "ugaz_price"})

print(ugaz3.shape)

ugaz3.head()
```

```
dgaz2 = dgaz.drop(['Open', 'High', 'Low', 'Adj Close', 'Volume'], axis=1)

dgaz3 = dgaz2.rename(index=str, columns={"Date": "date", "Close": "dgaz_price"})

print(dgaz3.shape)

dgaz3.head()
```

```
ng.isnull().sum()
```

```
ng = ng.set_index(['date'])
print(ng.loc['2018-01-04'])
```

```
merged = dgaz3.merge(ugaz3)

merged.head()
```

```
merged2 = merged.merge(ung4)
print(merged2.shape)
merged2.head()
```

```
merged3 = merged2.drop(merged2.index[1850:])

print(merged3.shape)

merged3.tail()
```

```
print(ng2.shape)
ng2.tail()
```

```
ng2 = ng2.astype(float)
```

```
df = pd.merge(ng2, merged3, how='outer', on='date')

print(df.dtypes)
print(df.shape)

df.head()
```

```
df.isnull().sum()
```

```
df2 = df.fillna(method='ffill')

df2.isnull().sum()
```

```
df2.tail()
```

```
df3 = df2.sort_values('date')

df3.tail()

# Final Table
```
